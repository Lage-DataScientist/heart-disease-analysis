## 1) Enquadramento Cl√≠nico: o que √© um Ataque Card√≠aco

Um **ataque card√≠aco** (MI) ocorre quando o **fluxo de sangue para uma parte do m√∫sculo card√≠aco √© bloqueado por tempo suficiente** para causar **les√£o ou morte do tecido**. A causa predominante √© a **aterosclerose**, isto √©, o **ac√∫mulo de placas** (gordura, colesterol e outras subst√¢ncias) nas **art√©rias coron√°rias** que irrigam o cora√ß√£o. Quando uma placa rompe e/ou se forma um trombo que **oclu√≠ a art√©ria**, o mioc√°rdio fica sem oxig√©nio (isquemia), podendo evoluir para necrose.

O cora√ß√£o contrai-se 60‚Äì80 vezes por minuto em repouso e, tal como os restantes √≥rg√£os, **precisa de suprimento sangu√≠neo pr√≥prio** atrav√©s das art√©rias coron√°rias. **Insufici√™ncia coron√°ria** surge quando esse suprimento √© comprometido. Clinicamente, isso manifesta-se num espectro que vai desde **angina est√°vel** (dor tor√°cica ao esfor√ßo, aliviada em repouso) at√© **s√≠ndromes coron√°rias agudas** (oclus√µes s√∫bitas com dor intensa e risco de morte s√∫bita).

## 3) Sintomas Cl√≠nicos

1. **Biologia e anatomia:**  
   - Mulheres t√™m art√©rias coron√°rias de menor calibre e podem desenvolver **doen√ßa microvascular**, mais dif√≠cil de detetar e com sintomas at√≠picos.

4. **Disparidades cl√≠nicas:**

**Implica√ß√£o cl√≠nica e para modelagem:** estas diferen√ßas refor√ßam a necessidade de **considerar o sexo como vari√°vel central nos modelos preditivos** e de avaliar m√©tricas separadas por g√©nero, de modo a reduzir atrasos diagn√≥sticos e aumentar a equidade.

### 1. Angina
A **angina** √© um tipo de dor tor√°cica provocada pela **redu√ß√£o do fluxo sangu√≠neo para o m√∫sculo card√≠aco**, sendo um dos sintomas mais t√≠picos da doen√ßa arterial coron√°ria.  

√â geralmente descrita como **press√£o, aperto, peso ou dor** no peito, podendo irradiar para ombros, bra√ßos, pesco√ßo, mand√≠bula ou costas. Existem v√°rios tipos de angina:  
- **Angina Est√°vel:** surge de forma previs√≠vel durante esfor√ßo f√≠sico ou stress e √© aliviada com repouso ou medica√ß√£o.  
- **Angina Inst√°vel:** ocorre de forma inesperada, podendo acontecer em repouso; √© mais grave e prolongada, configurando emerg√™ncia m√©dica.  
- **Angina Variante (Prinzmetal):** causada por **espasmos nas art√©rias coron√°rias**, surge geralmente em repouso e pode ser severa.  

> **Rela√ß√£o com o dataset:** a vari√°vel **cp (chest pain type)** codifica diferentes tipos de angina, permitindo distinguir entre padr√µes cl√≠nicos de risco vari√°vel.

### 2. Hipertrofia
**Hipertrofia** significa o **espessamento ou aumento do tamanho de um √≥rg√£o ou tecido**.  

No contexto card√≠aco, a **hipertrofia ventricular esquerda (LVH)** corresponde ao espessamento da parede do ventr√≠culo esquerdo, geralmente causada por **hipertens√£o arterial cr√≥nica** ou condi√ß√µes que aumentam a carga de trabalho do cora√ß√£o.  

A LVH est√° associada a complica√ß√µes como:  
- **Insufici√™ncia card√≠aca** (pelo esfor√ßo adicional imposto ao cora√ß√£o).  
- **Arritmias** (altera√ß√µes do ritmo card√≠aco).  
- **Doen√ßa isqu√©mica do cora√ß√£o.**

> **Rela√ß√£o com o dataset:** a vari√°vel **restecg (ECG em repouso)** pode indicar presen√ßa de hipertrofia ventricular, o que aumenta a probabilidade de eventos card√≠acos.

### 3. Segmento ST (ST Depression)
O **segmento ST** faz parte do ciclo el√©trico card√≠aco, registado no **eletrocardiograma (ECG)**.  

- Situa-se entre o final da onda S (√∫ltima deflex√£o do complexo QRS) e o in√≠cio da onda T.  
- Representa o per√≠odo de **despolariza√ß√£o ventricular**.  

**ST Depression** refere-se a um **deslocamento descendente do segmento ST** em rela√ß√£o √† linha de base.  
- √â um achado sugestivo de **isquemia mioc√°rdica** (falta de oxig√©nio no m√∫sculo card√≠aco), comum na **doen√ßa arterial coron√°ria**.  

> **Rela√ß√£o com o dataset:** vari√°veis como **oldpeak (depress√£o do ST induzida por exerc√≠cio)** e **slope (inclina√ß√£o do segmento ST)** s√£o preditores cl√≠nicos importantes de isquemia.

---

### 4. Fluoroscopia
A **fluoroscopia** √© uma t√©cnica de imagem que utiliza **raios-X em tempo real**, permitindo observar estruturas internas em movimento.  

No contexto da cardiologia, √© usada para **visualizar as art√©rias coron√°rias** ap√≥s administra√ß√£o de contraste, ajudando a identificar obstru√ß√µes ou estreitamentos.

### 5. Talassemia
A **talassemia** √© uma **doen√ßa gen√©tica do sangue** que compromete a produ√ß√£o de **hemoglobina** e de gl√≥bulos vermelhos.  

Consequ√™ncias cl√≠nicas:  
- **Anemia cr√≥nica**, com fadiga e fraqueza.  
- Potenciais complica√ß√µes graves em casos severos.  

Tipos principais:  
- **Talassemia Alfa:** muta√ß√µes nos genes da prote√≠na alfa globina.  
- **Talassemia Beta:** muta√ß√µes nos genes da prote√≠na beta globina.  

> **Rela√ß√£o com o dataset:** a vari√°vel **thal (thall)** classifica o estado associado √† talassemia (normal, defeito fixo, defeito revers√≠vel), que pode influenciar a oxigena√ß√£o e estar correlacionada com maior risco cardiovascular.

### 9.2 Testes de Normalidade

Foi criada a fun√ß√£o **`testar_normalidade`** para avaliar cada vari√°vel num√©rica em detalhe.  
Esta fun√ß√£o contempla os seguintes componentes:

- **Estat√≠sticas descritivas:** valores centrais (m√©dia, mediana, quartis) e medidas de dispers√£o (desvio padr√£o), complementados por **skewness** (assimetria) e **kurtosis** (achatamento).  
- **Formula√ß√£o de hip√≥teses:**
  - **H‚ÇÄ:** a vari√°vel segue distribui√ß√£o normal.  
  - **H‚ÇÅ:** a vari√°vel **n√£o** segue distribui√ß√£o normal.  
- **Testes aplicados:**  
  - *Shapiro-Wilk* (quando n < 5000).  
  - *D‚ÄôAgostino-Pearson (omnibus test)*.  
  - *Anderson-Darling*.  
- **Interpreta√ß√£o autom√°tica:** integra√ß√£o das conclus√µes dos testes para simplificar a leitura e apoiar a tomada de decis√£o.  
- **Visualiza√ß√£o gr√°fica:** histograma da vari√°vel com curva normal ajustada e **QQ-plot**, permitindo avaliar visualmente o ajustamento da distribui√ß√£o.

üìå **Decis√£o metodol√≥gica:**  
Aplicar **m√∫ltiplos testes de normalidade** aumenta a robustez da conclus√£o, evitando depender de um √∫nico teste que possa gerar falsos positivos/negativos.  
O uso de gr√°ficos complementa a an√°lise estat√≠stica, fornecendo uma avalia√ß√£o visual intuitiva.

---




````python
def mostrar_hipoteses(coluna):
    texto = f"""
<div style="text-align: center; font-size:16px; line-height:1.6">
    <h3>Teste de Normalidade para <code>{coluna}</code></h3>
    <b>H‚ÇÄ:</b> A vari√°vel <code>{coluna}</code> segue uma distribui√ß√£o normal <br>
    <b>H‚ÇÅ:</b> A vari√°vel <code>{coluna}</code> <u>n√£o</u> segue uma distribui√ß√£o normal
</div>
"""
    display(Markdown(texto))


def testar_normalidade(df, coluna, alpha=0.05, plot=True):
    """
    Testa normalidade de uma vari√°vel num√©rica.
    Inclui estat√≠sticas descritivas, hip√≥teses, testes e gr√°ficos.
    """
    serie = df[coluna].dropna()

    # -------------------------
    # üìä Estat√≠sticas descritivas
    stats_table = serie.describe().rename({
        'count': 'N¬∫ Registos',
        'mean': 'M√©dia',
        'std': 'Desvio Padr√£o',
        'min': 'M√≠nimo',
        '25%': '1¬∫ Quartil',
        '50%': 'Mediana',
        '75%': '3¬∫ Quartil',
        'max': 'M√°ximo'
    }).to_frame(name=coluna)

    stats_table["Assimetria"] = skew(serie)
    stats_table["Curtose"] = kurtosis(serie)

    display(HTML("<br>"))
    display(stats_table.style
            .format(precision=2, thousands=".", decimal=",")
            .set_caption("üìÅ Estat√≠sticas Descritivas")
            .set_table_attributes('style="width: 40%; margin: auto;"'))

    # -------------------------
    # üîé Mostrar hip√≥teses
    display(HTML("<br><br>"))
    mostrar_hipoteses(coluna)

    # -------------------------
    # üìë Testes de normalidade
    resultados = []

    if len(serie) <= 5000:
        stat, p = st.shapiro(serie)
        resultados.append(("Shapiro-Wilk", stat, p, "Rejeita H0" if p < alpha else "N√£o rejeita H0"))

    stat, p = st.normaltest(serie)
    resultados.append(("D‚ÄôAgostino-Pearson", stat, p, "Rejeita H0" if p < alpha else "N√£o rejeita H0"))

    ad = st.anderson(serie, dist="norm")
    critico = ad.critical_values[2]
    conclusao_ad = "Rejeita H0" if ad.statistic > critico else "N√£o rejeita H0"
    resultados.append(("Anderson-Darling", ad.statistic, np.nan, conclusao_ad))

    df_resultados = pd.DataFrame(resultados, columns=["Teste", "Estat√≠stica", "p-valor", "Conclus√£o"])

    display(HTML("<br><br>"))
    styled = (
        df_resultados.style
        .set_caption(f"üìä Testes de Normalidade para '{coluna}' (Œ±={alpha})")
        .set_table_styles(
            [
                {"selector": "caption", "props": [("caption-side", "top"), ("text-align", "center"),
                                                  ("font-size", "14px"), ("font-weight", "bold")]},
                {"selector": "table", "props": [("margin-left", "auto"), ("margin-right", "auto")]}
            ],
            overwrite=False
        )
    )
    html = styled.to_html()
    display(HTML(f"<div style='display: flex; justify-content: center;'>{html}</div>"))

    # -------------------------
    # üìñ Interpreta√ß√£o autom√°tica
    conclusoes = df_resultados["Conclus√£o"].value_counts()
    if "Rejeita H0" in conclusoes and conclusoes["Rejeita H0"] >= 2:
        interpretacao = f"‚ùå A vari√°vel '{coluna}' N√ÉO segue distribui√ß√£o normal (Rejeitamos H0 ao n√≠vel de {alpha})."
    else:
        interpretacao = f"‚úÖ A vari√°vel '{coluna}' segue distribui√ß√£o normal (N√£o rejeitamos H0 ao n√≠vel de {alpha})."

    print("\nüìñ Interpreta√ß√£o autom√°tica:")
    print(interpretacao)

    # -------------------------
    # üìà Gr√°ficos
    if plot:
        fig, axes = plt.subplots(1, 2, figsize=(12, 5))

        axes[0].hist(serie, bins=30, density=True, alpha=0.6, color='steelblue')
        mu, sigma = serie.mean(), serie.std()
        x = np.linspace(serie.min(), serie.max(), 100)
        axes[0].plot(x, st.norm.pdf(x, mu, sigma), 'r', linewidth=2)
        axes[0].set_title(f"Histograma de {coluna}")

        sm.qqplot(serie, line='s', ax=axes[1])
        axes[1].set_title(f"QQ-plot de {coluna}")

        plt.show()
````

### 9.3 An√°lise de Outliers

Foi criada a fun√ß√£o **`analisar_outliers_tukey`**, baseada no m√©todo das **cercas de Tukey (IQR method)**.  
A an√°lise segue os seguintes passos:

1. C√°lculo do **Q1 (1¬∫ quartil)**, **Q3 (3¬∫ quartil)** e do **IQR (Q3 - Q1)**.  
2. Determina√ß√£o dos **limites inferior e superior** (Q1 - k¬∑IQR e Q3 + k¬∑IQR, com k=1.5 por padr√£o).  
3. Identifica√ß√£o e contagem dos outliers.  
4. C√°lculo da percentagem de outliers em rela√ß√£o ao total.  
5. Visualiza√ß√£o atrav√©s de **boxplot horizontal**, com marca√ß√£o das cercas.

üìå **Decis√£o metodol√≥gica:**  
- Optou-se pelo m√©todo de **Tukey** por ser simples, robusto e amplamente utilizado em estat√≠stica aplicada.  
- Em dados cl√≠nicos, valores extremos **n√£o devem ser removidos automaticamente**, pois frequentemente representam **casos cr√≠ticos reais** (ex.: n√≠veis de colesterol >500 mg/dl).  
- A avalia√ß√£o de cada outlier ser√° feita **vari√°vel a vari√°vel**, considerando tanto a **plausibilidade m√©dica** quanto o **impacto estat√≠stico** na an√°lise e modelagem.

---

````python
def analisar_outliers_tukey(df, coluna, k=1.5, width=14, height=4):
    """
    Analisa outliers de uma vari√°vel num√©rica usando Tukey's fences (IQR).
    Mostra tabela resumo e boxplot horizontal com cercas.
    """
    serie = df[coluna].dropna()

    # Quartis e IQR
    Q1, Q3 = serie.quantile(0.25), serie.quantile(0.75)
    IQR = Q3 - Q1
    lower, upper = Q1 - k*IQR, Q3 + k*IQR

    # Outliers
    outliers = serie[(serie < lower) | (serie > upper)]

    #  Tabela com limites
    df_limites = pd.DataFrame({
        "Q1": [Q1],
        "Q3": [Q3],
        "IQR": [IQR],
        f"Limite Inferior (k={k})": [lower],
        f"Limite Superior (k={k})": [upper],
        "N¬∫ Outliers": [len(outliers)],
        "% Outliers": [len(outliers) / len(serie) * 100]
    }).T.rename(columns={0: "Valor"})

    display(HTML(f"<h3 style='text-align:center'>üìä An√°lise de Outliers (Tukey) para '{coluna}'</h3>"))
    display(df_limites.style.format({"Valor": "{:.2f}"}).set_table_styles(
        [{"selector": "table", "props": [("margin-left", "auto"), ("margin-right", "auto")]}]
    ))

    #  Boxplot horizontal
    plt.figure(figsize=(width, height))
    sns.boxplot(x=serie, orient="h", color="skyblue")
    plt.axvline(lower, color="green", linestyle="--", label="Cerca Inferior")
    plt.axvline(upper, color="red", linestyle="--", label="Cerca Superior")
    plt.title(f"Boxplot de {coluna} com Cercas de Tukey", fontsize=14, weight="bold")
    plt.legend(loc="upper right")
    plt.tight_layout()
    plt.show()

    return {"limite_inferior": lower, "limite_superior": upper, "outliers": outliers}

````

Estes registos s√£o raros, mas cr√≠ticos para a previs√£o de ataques card√≠acos e devem ser preservados integralmente na base de dados.


````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from IPython.display import display, HTML

def analisar_categorica_univariada(df, col, rename_dict=None, order=None, cores=None):
    """
    Analisa uma vari√°vel categ√≥rica:
    - Mostra tabela de frequ√™ncias absoluta e relativa (%)
    - Mostra gr√°fico de barras com valores e percentagens
    
    Par√¢metros:
    -----------
    df : DataFrame
        Base de dados
    col : str
        Nome da coluna categ√≥rica
    rename_dict : dict (opcional)
        Mapeamento de valores para r√≥tulos leg√≠veis (ex.: {1: "Homem", 0: "Mulher"})
    order : list (opcional)
        Ordem das categorias no gr√°fico
    cores : list (opcional)
        Lista de cores a aplicar √†s barras
    """
    
    # Criar s√©rie com r√≥tulos mais leg√≠veis (se houver dicion√°rio)
    serie = df[col].map(rename_dict) if rename_dict else df[col]
    
    # Calcular frequ√™ncias
    freq_table = serie.value_counts().reset_index()
    freq_table.columns = [col, 'Frequ√™ncia']
    freq_table['Percentagem (%)'] = 100 * freq_table['Frequ√™ncia'] / freq_table['Frequ√™ncia'].sum()
    
    # Mostrar tabela formatada e centrada
    display(freq_table.style
            .format({'Percentagem (%)': '{:.2f}%'})
            .set_caption(f"üìä Distribui√ß√£o de {col}")
            .set_table_attributes('style="margin:auto; width:50%; text-align:center;"'))
    display(HTML("<br><br>"))  # adiciona duas linhas em branco

    # Gr√°fico
    plt.figure(figsize=(8,5))
    ax = sns.countplot(x=serie, data=df, order=order, palette=cores)
    
    # Adicionar r√≥tulos nas barras
    total = len(df)
    for p in ax.patches:
        freq = p.get_height()
        perc = 100 * freq / total
        ax.annotate(f'{freq} ({perc:.1f}%)', 
                    (p.get_x() + p.get_width() / 2., freq), 
                    ha='center', va='bottom', fontsize=11, fontweight='bold')
    
    
    
    # Ajustes de estilo
    ax.set_title(f"Distribui√ß√£o de {col}", fontsize=14, weight="bold")
    ax.set_xlabel(col, fontsize=12)
    ax.set_ylabel("N√∫mero de Registos", fontsize=12)
    sns.despine()
    plt.show()
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy.stats import mannwhitneyu, pointbiserialr
from cliffs_delta import cliffs_delta
from IPython.display import display, HTML

def analisar_variavel(df, var, target="target"):
    """
    Faz an√°lise bivariada entre uma vari√°vel num√©rica e uma vari√°vel target bin√°ria (0/1).
    Mostra gr√°ficos, estat√≠sticas e interpreta√ß√£o autom√°tica com hip√≥teses expl√≠citas.
    """

    group0 = df.loc[df[target] == 0, var]
    group1 = df.loc[df[target] == 1, var]

    # ---------- Visualiza√ß√µes ----------
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    # Violin + Box
    sns.violinplot(x=df[target], y=df[var], palette="pastel", inner=None, ax=axes[0])
    sns.boxplot(x=df[target], y=df[var],
                width=0.2, boxprops={'facecolor':'none'}, ax=axes[0])
    axes[0].set_title(f"{var} vs {target} (Violin + Box)")
    axes[0].set_xlabel("Target")
    axes[0].set_ylabel(var)

    # KDE overlay
    sns.kdeplot(group0, label="Target=0", fill=True, ax=axes[1])
    sns.kdeplot(group1, label="Target=1", fill=True, ax=axes[1])
    axes[1].set_title(f"Distribui√ß√£o de {var} por {target}")
    axes[1].set_xlabel(var)
    axes[1].set_ylabel("Densidade")
    axes[1].legend()

    plt.tight_layout()
    plt.show()

    # ---------- Estat√≠sticas ----------
    desc_stats = df.groupby(target)[var].agg(['mean', 'median', 'std'])

    # Mann-Whitney U test
    u_stat, p_val = mannwhitneyu(group0, group1, alternative="two-sided")

    # Point-biserial correlation
    corr, p_corr = pointbiserialr(df[target], df[var])

    # Cliff‚Äôs Delta
    d, magnitude = cliffs_delta(group0, group1)

    # ---------- Resultados consolidados ----------
    resultados = pd.DataFrame({
        "M√©dia Grupo 0": [group0.mean()],
        "M√©dia Grupo 1": [group1.mean()],
        "Mediana Grupo 0": [group0.median()],
        "Mediana Grupo 1": [group1.median()],
        "Mann-Whitney U": [u_stat],
        "p-valor (U)": [p_val],
        "Correla√ß√£o r": [corr],
        "p-valor (r)": [p_corr],
        "Cliff‚Äôs Delta": [d],
        "Magnitude": [magnitude]
    })

    display(resultados.style.set_caption(f"üìä Resultados Bivariados para {var}")
            .format({
                "M√©dia Grupo 0": "{:.2f}",
                "M√©dia Grupo 1": "{:.2f}",
                "Mediana Grupo 0": "{:.2f}",
                "Mediana Grupo 1": "{:.2f}",
                "Mann-Whitney U": "{:.2f}",
                "p-valor (U)": "{:.4f}",
                "Correla√ß√£o r": "{:.3f}",
                "p-valor (r)": "{:.4f}",
                "Cliff‚Äôs Delta": "{:.3f}"
            }))

    # ---------- Interpreta√ß√£o autom√°tica (estilizada) ----------
    print("\n" + "="*60)
    print("üìñ Interpreta√ß√£o Autom√°tica".center(60))
    print("="*60)

    # Mann-Whitney
    print("\n‚û°Ô∏è Teste **Mann-Whitney U**")
    print("   ‚Ä¢ H‚ÇÄ: As distribui√ß√µes s√£o iguais entre target=0 e target=1")
    print("   ‚Ä¢ H‚ÇÅ: As distribui√ß√µes s√£o diferentes entre target=0 e target=1")
    if p_val < 0.05:
        print(f"   ‚úÖ Rejeitamos H‚ÇÄ (p={p_val:.4f}) ‚Üí **Diferen√ßa significativa entre grupos**")
    else:
        print(f"   ‚ùå N√£o rejeitamos H‚ÇÄ (p={p_val:.4f}) ‚Üí **Diferen√ßa n√£o significativa**")

    # Correla√ß√£o ponto-bisserial
    print("\n‚û°Ô∏è Teste **Correla√ß√£o Ponto-Bisserial**")
    print("   ‚Ä¢ H‚ÇÄ: N√£o existe correla√ß√£o linear entre vari√°vel e target (r=0)")
    print("   ‚Ä¢ H‚ÇÅ: Existe correla√ß√£o linear entre vari√°vel e target (r‚â†0)")
    if p_corr < 0.05:
        print(f"   ‚úÖ Rejeitamos H‚ÇÄ (p={p_corr:.4f}) ‚Üí **Correla√ß√£o significativa** (r={corr:.3f})")
    else:
        print(f"   ‚ùå N√£o rejeitamos H‚ÇÄ (p={p_corr:.4f}) ‚Üí **Correla√ß√£o n√£o significativa** (r={corr:.3f})")

    # Cliff‚Äôs Delta
    print("\n‚û°Ô∏è Medida de Efeito **Cliff‚Äôs Delta (Œ¥)**")
    print("   ‚Ä¢ N√£o √© um teste de hip√≥tese, mas uma medida da magnitude da diferen√ßa")
    print(f"   ‚öñÔ∏è Œ¥ = {d:.3f} ‚Üí **{magnitude} efeito**")
    print("="*60 + "\n")

````

Apesar da aparente invers√£o cl√≠nica (maiores valores em n√£o doentes), √© uma das vari√°veis mais informativas do dataset, devendo ser considerada com aten√ß√£o em an√°lises preditivas e interpretada √† luz do contexto cl√≠nico.


````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency, fisher_exact
from IPython.display import display, HTML

# Fun√ß√£o para calcular Cram√©r's V
def cramers_v(confusion_matrix):
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    r, k = confusion_matrix.shape
    return np.sqrt(chi2 / (n * (min(k, r)-1)))

def analisar_categorica(df, var, target="target", alpha=0.05, export=False,
                        rename_dict=None, rename_dict_target=None):
    """
    An√°lise bivariada entre uma vari√°vel categ√≥rica e um target bin√°rio/categ√≥rico.
    Inclui:
        - Visualiza√ß√£o (countplot com contagens + percentuais nas barras)
        - Tabelas de conting√™ncia (absoluta e relativa)
        - Testes Qui-quadrado e Fisher (2x2)
        - Medida de efeito (Cram√©r‚Äôs V)
        - Resumo autom√°tico interpretado
    """

    print(f"\nüìä Resultados para {var}")
    print("-" * 70)

    df = df.copy()

    # ---------- Renomear categorias ----------
    if rename_dict:
        df[var] = df[var].map(rename_dict).fillna(df[var])

    if rename_dict_target:
        df[target] = df[target].map(rename_dict_target).fillna(df[target])

    # ---------- Visualiza√ß√£o ----------
    plt.figure(figsize=(10, 6))
    ax = sns.countplot(x=df[var], hue=df[target], palette="pastel")

    # Calcular percentagens dentro de cada categoria (j√° renomeada)
    group_totals = df.groupby(df[var])[target].count()
    group_totals.index = group_totals.index.astype(str)  # garante consist√™ncia de tipos

    for container in ax.containers:
        labels = []
        for rect in container:
            height = int(rect.get_height())
            if height > 0:
                # categoria renomeada no eixo X
                cat_idx = int(rect.get_x() + rect.get_width() / 2.0 + 0.5)
                if cat_idx < len(ax.get_xticklabels()):
                    category = ax.get_xticklabels()[cat_idx].get_text()
                    # usa .get() para evitar KeyError
                    pct = 100 * height / group_totals.get(str(category), height)
                    labels.append(f"{height}\n({pct:.1f}%)")
                else:
                    labels.append("")
            else:
                labels.append("")
        ax.bar_label(container, labels=labels, label_type="edge", fontsize=9)

    # Legenda ajustada
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles, labels, title=target, loc="upper right")

    plt.title(f'{var} vs. {target}')
    plt.xlabel(var)
    plt.ylabel("Contagem")
    plt.tight_layout()
    plt.show()

    # ---------- Tabelas ----------
    contingency = pd.crosstab(df[var], df[target])

    display(HTML("<div style='text-align:center'>"))
    display(contingency.style
            .set_caption("üìå Tabela de Conting√™ncia (absoluta)")
            .set_table_attributes('style="margin:auto"')
            .background_gradient(axis=None, cmap="Blues"))

    contingency_pct = contingency.apply(lambda x: x / x.sum(), axis=1)
    display(contingency_pct.style
            .set_caption("üìå Tabela de Conting√™ncia (% por linha)")
            .set_table_attributes('style="margin:auto"')
            .format("{:.2%}")
            .background_gradient(axis=None, cmap="Greens"))
    display(HTML("</div>"))

    # ---------- Testes de hip√≥teses ----------
    chi2, p, dof, expected = chi2_contingency(contingency)

    fisher_res = None
    if contingency.shape == (2, 2):
        oddsratio, p_fisher = fisher_exact(contingency)
        fisher_res = (oddsratio, p_fisher)

    # ---------- Medida de efeito ----------
    v = cramers_v(contingency)
    if v < 0.1:
        interpret = "negligenci√°vel"
    elif v < 0.3:
        interpret = "pequena"
    elif v < 0.5:
        interpret = "moderada"
    else:
        interpret = "forte"

    # ---------- Resumo autom√°tico ----------
    print("\n" + "="*60)
    print(f"üìñ Interpreta√ß√£o Autom√°tica de {var}".center(60))
    print("="*60)

    print("\n‚û°Ô∏è Teste Qui-quadrado de independ√™ncia")
    print("   ‚Ä¢ H‚ÇÄ: N√£o existe associa√ß√£o entre as vari√°veis (independ√™ncia)")
    print("   ‚Ä¢ H‚ÇÅ: Existe associa√ß√£o entre as vari√°veis (depend√™ncia)")
    if p < alpha:
        print(f"   ‚úÖ Rejeitamos H‚ÇÄ (p={p:.4f}) ‚Üí **Associa√ß√£o significativa**")
    else:
        print(f"   ‚ùå N√£o rejeitamos H‚ÇÄ (p={p:.4f}) ‚Üí **Sem evid√™ncia de associa√ß√£o**")

    if fisher_res:
        oddsratio, p_fisher = fisher_res
        print("\n‚û°Ô∏è Teste Exato de Fisher (2x2)")
        print("   ‚Ä¢ H‚ÇÄ: As probabilidades de ocorr√™ncia do target s√£o iguais entre categorias")
        print("   ‚Ä¢ H‚ÇÅ: As probabilidades diferem entre categorias")
        if p_fisher < alpha:
            print(f"   ‚úÖ Rejeitamos H‚ÇÄ (p={p_fisher:.4f}) ‚Üí **Associa√ß√£o significativa** (OR={oddsratio:.2f})")
        else:
            print(f"   ‚ùå N√£o rejeitamos H‚ÇÄ (p={p_fisher:.4f}) ‚Üí **Sem evid√™ncia de associa√ß√£o** (OR={oddsratio:.2f})")

    print("\n‚û°Ô∏è Medida de efeito: Cram√©r‚Äôs V")
    print(f"   ‚Ä¢ V = {v:.3f} ‚Üí **{interpret} associa√ß√£o**")
    print("="*60 + "\n")

    # ---------- Exporta√ß√£o ----------
    if export:
        resultados = {
            "Vari√°vel": var,
            "Chi2": chi2,
            "p-valor Qui2": p,
            "Cram√©r's V": v,
            "For√ßa": interpret
        }
        if fisher_res:
            resultados.update({"OddsRatio": oddsratio, "p-valor Fisher": p_fisher})
        return resultados
````

### üîé Justifica√ß√£o Metodol√≥gica ‚Äî Correla√ß√£o (Spearman) e VIF

#### 1. Porque utiliz√°mos a correla√ß√£o de Spearman (e n√£o a de Pearson)
A an√°lise de correla√ß√£o foi realizada com o coeficiente **de Spearman** em vez do coeficiente **de Pearson**, pelas seguintes raz√µes:

- **Suposi√ß√µes menos restritivas:**  
  O coeficiente de Pearson assume linearidade e distribui√ß√£o normal das vari√°veis.  
  J√° Spearman, por ser uma correla√ß√£o baseada em **ranks**, n√£o depende da normalidade nem da linearidade estrita.  

- **Robustez a outliers e assimetrias:**  
  Muitas das vari√°veis num√©ricas do dataset (ex.: colesterol, press√£o arterial, oldpeak) apresentam **assimetria** e **outliers identificados nas an√°lises univariadas**.  
  O coeficiente de Spearman √© mais robusto a esses desvios, tornando-se mais adequado.  

- **Natureza dos dados cl√≠nicos:**  
  Vari√°veis fisiol√≥gicas nem sempre t√™m rela√ß√£o linear com a presen√ßa de doen√ßa, mas podem apresentar rela√ß√µes **monot√≥nicas** (√† medida que a vari√°vel aumenta, o risco pode aumentar ou diminuir, mas n√£o necessariamente de forma linear).  
  O Spearman capta essas tend√™ncias monot√≥nicas, ao contr√°rio do Pearson.

**Conclus√£o:** A escolha do Spearman garante uma medida de associa√ß√£o mais fi√°vel e robusta para este tipo de dados cl√≠nicos.

---

#### 2. Porque complement√°mos com a an√°lise de VIF
Ap√≥s a an√°lise da matriz de correla√ß√£o de Spearman, foi efetuada a an√°lise do **Variance Inflation Factor (VIF)** com o objetivo de avaliar a **multicolinearidade** entre as vari√°veis explicativas.

- **Limita√ß√£o da matriz de correla√ß√£o:**  
  A matriz de correla√ß√£o apenas avalia rela√ß√µes **bivariadas** (entre dois pares de vari√°veis).  
  Uma vari√°vel pode n√£o ter correla√ß√£o alta com nenhuma outra em particular, mas ainda assim ser explicada por uma **combina√ß√£o linear de v√°rias vari√°veis**.  

- **VIF como an√°lise multivariada:**  
  O VIF mede quanto a vari√¢ncia de cada vari√°vel √© inflacionada pela presen√ßa das outras no modelo.  
  Valores elevados de VIF (‚â• 5 ou ‚â• 10, consoante a literatura) indicam **colinearidade preocupante**, o que pode:  
  - Inflacionar erros-padr√£o,  
  - Instabilizar coeficientes de regress√£o,  
  - Dificultar a interpreta√ß√£o da contribui√ß√£o individual de cada vari√°vel.

- **Import√¢ncia para modelagem:**  
  Em modelos preditivos (como regress√µes log√≠sticas ou lineares), a multicolinearidade compromete a **estabilidade e interpretabilidade**.  
  A an√°lise do VIF assegura que as vari√°veis retidas s√£o estatisticamente consistentes.

**Conclus√£o:**  
A matriz de correla√ß√£o de Spearman permitiu identificar **associa√ß√µes monot√≥nicas e robustas** entre vari√°veis.  
O VIF complementou esta an√°lise ao detetar redund√¢ncias **multivariadas** no conjunto completo, garantindo que as vari√°veis selecionadas para os modelos n√£o apresentam colinearidade problem√°tica.

````python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd
import statsmodels.api as sm
from IPython.display import display

def calcular_vif_final(df, numeric_var, categoric_var):
    # 1. Selecionar num√©ricas
    num_data = df[numeric_var].dropna()
    
    # 2. Criar dummies para categ√≥ricas
    cat_data = pd.get_dummies(df[categoric_var], drop_first=True)
    
    # 3. Juntar num√©ricas + categ√≥ricas
    X_all = pd.concat([num_data, cat_data], axis=1)
    X_all = sm.add_constant(X_all)
    
    # 4. Calcular VIF
    vif_all = pd.DataFrame()
    vif_all["Vari√°vel"] = X_all.columns
    vif_all["VIF"] = [variance_inflation_factor(X_all.values, i) for i in range(X_all.shape[1])]
    
    # 5. Mostrar tabela estilizada
    display(
        vif_all.sort_values("VIF", ascending=False).style
        .set_caption(" Fatores de Infla√ß√£o da Vari√¢ncia (VIF) ‚Äî Todas as Vari√°veis")
        .format({"VIF": "{:.2f}"})
        .background_gradient(subset="VIF", cmap="coolwarm")
        .set_table_styles(
            [
                {"selector": "caption", "props": [("font-size", "16px"), ("font-weight", "bold"), ("color", "#9EA3B1")]},
                {"selector": "th", "props": [("background-color", "#1F1F1F"), ("font-weight", "bold")]}
            ]
        )
    )
    
    return vif_all.sort_values("VIF", ascending=False)

````

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_validate
from sklearn.preprocessing import RobustScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    roc_auc_score,
    roc_curve
)

# ======================
# 1. Definir X e y
# ======================
X = df.drop(columns=["target"])
y = df["target"]

# ======================
# 2. Identificar colunas
# ======================
numeric_features = ["age", "trtbps", "chol", "thalach", "oldpeak"]
categorical_features = ["sex", "cp", "fbs", "rest_ecg", "exang", "slope", "ca", "thal"]

# ======================
# 3. Pr√©-processamento
# ======================
preprocessor = ColumnTransformer(
    transformers=[
        ("num", RobustScaler(), numeric_features),
        ("cat", OneHotEncoder(drop="first"), categorical_features)
    ]
)

# ======================
# 4. Criar pipeline
# ======================
pipe = Pipeline([
    ("preprocessor", preprocessor),
    ("log_reg", LogisticRegression(max_iter=1000, random_state=19))
])

# ======================
# 5. Divis√£o treino/teste
# ======================
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=3, stratify=y
)

import optuna
import numpy as np
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

# =============================
# Fun√ß√£o objetivo do Optuna
# =============================
def objective(trial):
    model_type = trial.suggest_categorical("model", [
        "logreg", "rf", "xgb", "lgbm", "catboost", "knn", "svc"
    ])

    if model_type == "logreg":
        C = trial.suggest_loguniform("C", 1e-3, 1e3)
        penalty = trial.suggest_categorical("penalty", ["l1", "l2"])
        solver = "liblinear" if penalty == "l1" else "lbfgs"
        model = LogisticRegression(
            C=C, penalty=penalty, solver=solver, max_iter=500
        )

    elif model_type == "rf":
        n_estimators = trial.suggest_int("n_estimators", 100, 500)
        max_depth = trial.suggest_int("max_depth", 3, 15)
        min_samples_split = trial.suggest_int("min_samples_split", 2, 10)
        min_samples_leaf = trial.suggest_int("min_samples_leaf", 1, 5)
        model = RandomForestClassifier(
            n_estimators=n_estimators,
            max_depth=max_depth,
            min_samples_split=min_samples_split,
            min_samples_leaf=min_samples_leaf,
            random_state=3
        )

    elif model_type == "xgb":
        n_estimators = trial.suggest_int("n_estimators", 100, 500)
        max_depth = trial.suggest_int("max_depth", 3, 10)
        learning_rate = trial.suggest_loguniform("learning_rate", 0.01, 0.3)
        subsample = trial.suggest_uniform("subsample", 0.6, 1.0)
        colsample_bytree = trial.suggest_uniform("colsample_bytree", 0.6, 1.0)
        model = XGBClassifier(
            n_estimators=n_estimators,
            max_depth=max_depth,
            learning_rate=learning_rate,
            subsample=subsample,
            colsample_bytree=colsample_bytree,
            random_state=3,
            use_label_encoder=False,
            eval_metric="logloss"
        )

    elif model_type == "lgbm":
        n_estimators = trial.suggest_int("n_estimators", 100, 500)
        max_depth = trial.suggest_int("max_depth", -1, 15)  # -1 = sem limite
        learning_rate = trial.suggest_loguniform("learning_rate", 0.01, 0.3)
        num_leaves = trial.suggest_int("num_leaves", 15, 50)
        model = LGBMClassifier(
            n_estimators=n_estimators,
            max_depth=max_depth,
            learning_rate=learning_rate,
            num_leaves=num_leaves,
            random_state=3
        )

    elif model_type == "catboost":
        depth = trial.suggest_int("depth", 3, 10)
        learning_rate = trial.suggest_loguniform("learning_rate", 0.01, 0.3)
        iterations = trial.suggest_int("iterations", 100, 500)
        model = CatBoostClassifier(
            depth=depth,
            learning_rate=learning_rate,
            iterations=iterations,
            random_state=3,
            verbose=0
        )

    elif model_type == "knn":
        n_neighbors = trial.suggest_int("n_neighbors", 3, 15)
        metric = trial.suggest_categorical("metric", ["euclidean", "manhattan", "minkowski"])
        model = KNeighborsClassifier(
            n_neighbors=n_neighbors, metric=metric
        )

    elif model_type == "svc":
        C = trial.suggest_loguniform("C", 1e-3, 1e3)
        kernel = trial.suggest_categorical("kernel", ["linear", "rbf", "poly"])
        gamma = trial.suggest_loguniform("gamma", 1e-4, 1e-1) if kernel != "linear" else "scale"
        model = SVC(
            C=C, kernel=kernel, gamma=gamma, probability=True
        )

    # =============================
    # Valida√ß√£o cruzada estratificada
    # =============================
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=3)
    scores = cross_val_score(model, X_train, y_train, cv=cv, scoring="roc_auc")
    
    return np.mean(scores)

# =============================
# Executar estudo
# =============================
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=100, n_jobs=-1)  # 100 itera√ß√µes

print("Melhor modelo:", study.best_trial.params["model"])
print("Melhores par√¢metros:", study.best_trial.params)
print("Melhor ROC AUC:", study.best_value)
````

````python
import numpy as np

class ThresholdClassifier:
    def __init__(self, pipeline, threshold=0.5):
        self.pipeline = pipeline
        self.threshold = threshold
    
    def fit(self, X, y):
        self.pipeline.fit(X, y)
        return self
    
    def predict_proba(self, X):
        return self.pipeline.predict_proba(X)
    
    def predict(self, X):
        probs = self.predict_proba(X)[:, 1]  # probabilidade da classe positiva
        return (probs >= self.threshold).astype(int)
````

````python
import shap

class ThresholdClassifierExplainable(ThresholdClassifier):
    def explain_prediction(self, X, top_n=3):
        """
        Devolve a probabilidade, classe predita e top N vari√°veis mais influentes
        """
        # Probabilidade e classe
        prob = self.predict_proba(X)[:, 1][0]
        classe = "Doente" if prob >= self.threshold else "Saud√°vel"
        
        # SHAP explainer (para modelos baseados em √°rvores √© r√°pido)
        explainer = shap.TreeExplainer(self.pipeline.named_steps["classifier"])
        
        # ‚ö†Ô∏è Aten√ß√£o: o explainer precisa dos dados j√° transformados
        X_trans = self.pipeline.named_steps["preprocessor"].transform(X)
        shap_values = explainer.shap_values(X_trans)
        
        # Se binary, shap_values √© lista [classe0, classe1]
        if isinstance(shap_values, list):
            shap_values = shap_values[1]
        
        # Obter nomes das vari√°veis
        feature_names_num = numeric_features
        feature_names_cat = self.pipeline.named_steps["preprocessor"].transformers_[1][1]\
            .get_feature_names_out(categorical_features)
        feature_names = np.concatenate([feature_names_num, feature_names_cat])
        
        # SHAP para o paciente em quest√£o
        shap_patient = shap_values[0]  # primeira linha
        
        # Ordenar por import√¢ncia absoluta
        idx_sorted = np.argsort(np.abs(shap_patient))[::-1][:top_n]
        top_features = [(feature_names[i], shap_patient[i]) for i in idx_sorted]
        
        return {
            "Probabilidade": prob,
            "Classe": classe,
            "Top_Variaveis": top_features
        }
````
