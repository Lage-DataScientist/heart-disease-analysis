## 1) Enquadramento Clínico: o que é um Ataque Cardíaco

Um **ataque cardíaco** (MI) ocorre quando o **fluxo de sangue para uma parte do músculo cardíaco é bloqueado por tempo suficiente** para causar **lesão ou morte do tecido**. A causa predominante é a **aterosclerose**, isto é, o **acúmulo de placas** (gordura, colesterol e outras substâncias) nas **artérias coronárias** que irrigam o coração. Quando uma placa rompe e/ou se forma um trombo que **ocluí a artéria**, o miocárdio fica sem oxigénio (isquemia), podendo evoluir para necrose.

O coração contrai-se 60–80 vezes por minuto em repouso e, tal como os restantes órgãos, **precisa de suprimento sanguíneo próprio** através das artérias coronárias. **Insuficiência coronária** surge quando esse suprimento é comprometido. Clinicamente, isso manifesta-se num espectro que vai desde **angina estável** (dor torácica ao esforço, aliviada em repouso) até **síndromes coronárias agudas** (oclusões súbitas com dor intensa e risco de morte súbita).

## 2) Fatores Médicos e Fisiológicos Associados

### Causas e Fatores de Risco
- **Aterosclerose:** mecanismo principal — **estreitamento/oclusão** das coronárias por placas ateromatosas.  
- **Idade:** risco cresce com o envelhecimento.  
- **Sexo:** homens apresentam maior risco em idades mais precoces; após menopausa o risco nas mulheres aproxima-se.  
- **História familiar/genética:** predisposição a doença cardiovascular.  
- **Tabagismo:** lesão endotelial e aceleração da formação de placas.  
- **Hipertensão arterial:** aumenta pós-carga e dano vascular.  
- **Dislipidemia:** LDL elevado favorece aterogénese.  
- **Diabetes:** hiperglicemia crónica danifica vasos.  
- **Obesidade:** associa-se a HTA, diabetes e inflamação crónica.  
- **Sedentarismo:** piora perfil metabólico.  
- **Dieta inadequada:** excesso de gorduras saturadas/trans e colesterol.  
- **Stress crónico:** desregulação autonómica e inflamatória.  
- **Álcool (excesso):** aumenta PA e risco cardiovascular.

## 3) Sintomas Clínicos

### Sintomas Gerais
- **Dor ou desconforto torácico:** sensação de pressão ou aperto.  
- **Dor irradiada:** braços, costas, pescoço, mandíbula ou estômago.  
- **Falta de ar:** pode ocorrer com ou sem dor torácica.  
- **Sintomas associados:** sudorese fria, náuseas, tonturas ou desmaio.  

### Sintomas Comuns em Homens
- **Dor ou desconforto torácico:** geralmente descrito como aperto intenso no centro do peito.  
- **Dor irradiada:** frequentemente para o braço esquerdo, mas também ombro, costas ou mandíbula.  
- **Falta de ar:** comum, geralmente acompanhando dor torácica.  
- **Sintomas vegetativos:** suor frio, náusea, tontura.  
- **Padrão súbito e intenso:** início abrupto, sintomas marcantes e dramáticos.  

### Sintomas Comuns em Mulheres
- **Dor torácica menos típica:** aperto difuso ou pressão, não necessariamente intensa.  
- **Irradiação variável:** dor em ambos os braços, pescoço, costas ou mandíbula.  
- **Falta de ar:** pode surgir isoladamente.  
- **Sintomas inespecíficos:** fadiga incomum, indigestão, ansiedade, distúrbios do sono.  
- **Náuseas e vómitos:** mais prevalentes em mulheres.  
- **Padrão gradual ou intermitente:** sintomas podem ser subtis, intermitentes e facilmente confundidos.  

### Porque existem diferenças entre homens e mulheres?
1. **Biologia e anatomia:**  
   - Mulheres têm artérias coronárias de menor calibre e podem desenvolver **doença microvascular**, mais difícil de detetar e com sintomas atípicos.  
2. **Influência hormonal:**  
   - O estrogénio protege parcialmente contra doença coronária antes da menopausa. Após a menopausa, o risco e a manifestação alteram-se.  
3. **Perceção da dor e fatores psicossociais:**  
   - Mulheres tendem a reportar dor de forma diferente, aumentando o risco de **subdiagnóstico**.  
4. **Disparidades clínicas:**  
   - Protocolos de diagnóstico historicamente validados em homens podem não captar sintomas atípicos em mulheres.  

**Implicação clínica e para modelagem:** estas diferenças reforçam a necessidade de **considerar o sexo como variável central nos modelos preditivos** e de avaliar métricas separadas por género, de modo a reduzir atrasos diagnósticos e aumentar a equidade.

## 7) Explicação de Termos Técnicos Relacionados com as Variáveis

Ao longo do dataset surgem variáveis com terminologia médica específica. Abaixo descrevemos de forma detalhada os principais conceitos, com relevância direta para a interpretação clínica e para a análise preditiva.

---

### 1. Angina
A **angina** é um tipo de dor torácica provocada pela **redução do fluxo sanguíneo para o músculo cardíaco**, sendo um dos sintomas mais típicos da doença arterial coronária.  

É geralmente descrita como **pressão, aperto, peso ou dor** no peito, podendo irradiar para ombros, braços, pescoço, mandíbula ou costas. Existem vários tipos de angina:  
- **Angina Estável:** surge de forma previsível durante esforço físico ou stress e é aliviada com repouso ou medicação.  
- **Angina Instável:** ocorre de forma inesperada, podendo acontecer em repouso; é mais grave e prolongada, configurando emergência médica.  
- **Angina Variante (Prinzmetal):** causada por **espasmos nas artérias coronárias**, surge geralmente em repouso e pode ser severa.  

> **Relação com o dataset:** a variável **cp (chest pain type)** codifica diferentes tipos de angina, permitindo distinguir entre padrões clínicos de risco variável.

---

### 2. Hipertrofia
**Hipertrofia** significa o **espessamento ou aumento do tamanho de um órgão ou tecido**.  

No contexto cardíaco, a **hipertrofia ventricular esquerda (LVH)** corresponde ao espessamento da parede do ventrículo esquerdo, geralmente causada por **hipertensão arterial crónica** ou condições que aumentam a carga de trabalho do coração.  

A LVH está associada a complicações como:  
- **Insuficiência cardíaca** (pelo esforço adicional imposto ao coração).  
- **Arritmias** (alterações do ritmo cardíaco).  
- **Doença isquémica do coração.**

> **Relação com o dataset:** a variável **restecg (ECG em repouso)** pode indicar presença de hipertrofia ventricular, o que aumenta a probabilidade de eventos cardíacos.

---

### 3. Segmento ST (ST Depression)
O **segmento ST** faz parte do ciclo elétrico cardíaco, registado no **eletrocardiograma (ECG)**.  

- Situa-se entre o final da onda S (última deflexão do complexo QRS) e o início da onda T.  
- Representa o período de **despolarização ventricular**.  

**ST Depression** refere-se a um **deslocamento descendente do segmento ST** em relação à linha de base.  
- É um achado sugestivo de **isquemia miocárdica** (falta de oxigénio no músculo cardíaco), comum na **doença arterial coronária**.  

> **Relação com o dataset:** variáveis como **oldpeak (depressão do ST induzida por exercício)** e **slope (inclinação do segmento ST)** são preditores clínicos importantes de isquemia.

---

### 4. Fluoroscopia
A **fluoroscopia** é uma técnica de imagem que utiliza **raios-X em tempo real**, permitindo observar estruturas internas em movimento.  

No contexto da cardiologia, é usada para **visualizar as artérias coronárias** após administração de contraste, ajudando a identificar obstruções ou estreitamentos.  

> **Relação com o dataset:** a variável **ca (caa)** indica o número de **vasos principais visualizados por fluoroscopia**, refletindo diretamente o grau de doença coronária estrutural.

---

### 5. Talassemia
A **talassemia** é uma **doença genética do sangue** que compromete a produção de **hemoglobina** e de glóbulos vermelhos.  

Consequências clínicas:  
- **Anemia crónica**, com fadiga e fraqueza.  
- Potenciais complicações graves em casos severos.  

Tipos principais:  
- **Talassemia Alfa:** mutações nos genes da proteína alfa globina.  
- **Talassemia Beta:** mutações nos genes da proteína beta globina.  

> **Relação com o dataset:** a variável **thal (thall)** classifica o estado associado à talassemia (normal, defeito fixo, defeito reversível), que pode influenciar a oxigenação e estar correlacionada com maior risco cardiovascular.

### 9.2 Testes de Normalidade

Foi criada a função **`testar_normalidade`** para avaliar cada variável numérica em detalhe.  
Esta função contempla os seguintes componentes:

- **Estatísticas descritivas:** valores centrais (média, mediana, quartis) e medidas de dispersão (desvio padrão), complementados por **skewness** (assimetria) e **kurtosis** (achatamento).  
- **Formulação de hipóteses:**
  - **H₀:** a variável segue distribuição normal.  
  - **H₁:** a variável **não** segue distribuição normal.  
- **Testes aplicados:**  
  - *Shapiro-Wilk* (quando n < 5000).  
  - *D’Agostino-Pearson (omnibus test)*.  
  - *Anderson-Darling*.  
- **Interpretação automática:** integração das conclusões dos testes para simplificar a leitura e apoiar a tomada de decisão.  
- **Visualização gráfica:** histograma da variável com curva normal ajustada e **QQ-plot**, permitindo avaliar visualmente o ajustamento da distribuição.

📌 **Decisão metodológica:**  
Aplicar **múltiplos testes de normalidade** aumenta a robustez da conclusão, evitando depender de um único teste que possa gerar falsos positivos/negativos.  
O uso de gráficos complementa a análise estatística, fornecendo uma avaliação visual intuitiva.

---




````python
def mostrar_hipoteses(coluna):
    texto = f"""
<div style="text-align: center; font-size:16px; line-height:1.6">
    <h3>Teste de Normalidade para <code>{coluna}</code></h3>
    <b>H₀:</b> A variável <code>{coluna}</code> segue uma distribuição normal <br>
    <b>H₁:</b> A variável <code>{coluna}</code> <u>não</u> segue uma distribuição normal
</div>
"""
    display(Markdown(texto))


def testar_normalidade(df, coluna, alpha=0.05, plot=True):
    """
    Testa normalidade de uma variável numérica.
    Inclui estatísticas descritivas, hipóteses, testes e gráficos.
    """
    serie = df[coluna].dropna()

    # -------------------------
    # 📊 Estatísticas descritivas
    stats_table = serie.describe().rename({
        'count': 'Nº Registos',
        'mean': 'Média',
        'std': 'Desvio Padrão',
        'min': 'Mínimo',
        '25%': '1º Quartil',
        '50%': 'Mediana',
        '75%': '3º Quartil',
        'max': 'Máximo'
    }).to_frame(name=coluna)

    stats_table["Assimetria"] = skew(serie)
    stats_table["Curtose"] = kurtosis(serie)

    display(HTML("<br>"))
    display(stats_table.style
            .format(precision=2, thousands=".", decimal=",")
            .set_caption("📁 Estatísticas Descritivas")
            .set_table_attributes('style="width: 40%; margin: auto;"'))

    # -------------------------
    # 🔎 Mostrar hipóteses
    display(HTML("<br><br>"))
    mostrar_hipoteses(coluna)

    # -------------------------
    # 📑 Testes de normalidade
    resultados = []

    if len(serie) <= 5000:
        stat, p = st.shapiro(serie)
        resultados.append(("Shapiro-Wilk", stat, p, "Rejeita H0" if p < alpha else "Não rejeita H0"))

    stat, p = st.normaltest(serie)
    resultados.append(("D’Agostino-Pearson", stat, p, "Rejeita H0" if p < alpha else "Não rejeita H0"))

    ad = st.anderson(serie, dist="norm")
    critico = ad.critical_values[2]
    conclusao_ad = "Rejeita H0" if ad.statistic > critico else "Não rejeita H0"
    resultados.append(("Anderson-Darling", ad.statistic, np.nan, conclusao_ad))

    df_resultados = pd.DataFrame(resultados, columns=["Teste", "Estatística", "p-valor", "Conclusão"])

    display(HTML("<br><br>"))
    styled = (
        df_resultados.style
        .set_caption(f"📊 Testes de Normalidade para '{coluna}' (α={alpha})")
        .set_table_styles(
            [
                {"selector": "caption", "props": [("caption-side", "top"), ("text-align", "center"),
                                                  ("font-size", "14px"), ("font-weight", "bold")]},
                {"selector": "table", "props": [("margin-left", "auto"), ("margin-right", "auto")]}
            ],
            overwrite=False
        )
    )
    html = styled.to_html()
    display(HTML(f"<div style='display: flex; justify-content: center;'>{html}</div>"))

    # -------------------------
    # 📖 Interpretação automática
    conclusoes = df_resultados["Conclusão"].value_counts()
    if "Rejeita H0" in conclusoes and conclusoes["Rejeita H0"] >= 2:
        interpretacao = f"❌ A variável '{coluna}' NÃO segue distribuição normal (Rejeitamos H0 ao nível de {alpha})."
    else:
        interpretacao = f"✅ A variável '{coluna}' segue distribuição normal (Não rejeitamos H0 ao nível de {alpha})."

    print("\n📖 Interpretação automática:")
    print(interpretacao)

    # -------------------------
    # 📈 Gráficos
    if plot:
        fig, axes = plt.subplots(1, 2, figsize=(12, 5))

        axes[0].hist(serie, bins=30, density=True, alpha=0.6, color='steelblue')
        mu, sigma = serie.mean(), serie.std()
        x = np.linspace(serie.min(), serie.max(), 100)
        axes[0].plot(x, st.norm.pdf(x, mu, sigma), 'r', linewidth=2)
        axes[0].set_title(f"Histograma de {coluna}")

        sm.qqplot(serie, line='s', ax=axes[1])
        axes[1].set_title(f"QQ-plot de {coluna}")

        plt.show()
````

### 9.3 Análise de Outliers

Foi criada a função **`analisar_outliers_tukey`**, baseada no método das **cercas de Tukey (IQR method)**.  
A análise segue os seguintes passos:

1. Cálculo do **Q1 (1º quartil)**, **Q3 (3º quartil)** e do **IQR (Q3 - Q1)**.  
2. Determinação dos **limites inferior e superior** (Q1 - k·IQR e Q3 + k·IQR, com k=1.5 por padrão).  
3. Identificação e contagem dos outliers.  
4. Cálculo da percentagem de outliers em relação ao total.  
5. Visualização através de **boxplot horizontal**, com marcação das cercas.

📌 **Decisão metodológica:**  
- Optou-se pelo método de **Tukey** por ser simples, robusto e amplamente utilizado em estatística aplicada.  
- Em dados clínicos, valores extremos **não devem ser removidos automaticamente**, pois frequentemente representam **casos críticos reais** (ex.: níveis de colesterol >500 mg/dl).  
- A avaliação de cada outlier será feita **variável a variável**, considerando tanto a **plausibilidade médica** quanto o **impacto estatístico** na análise e modelagem.

---

````python
def analisar_outliers_tukey(df, coluna, k=1.5, width=14, height=4):
    """
    Analisa outliers de uma variável numérica usando Tukey's fences (IQR).
    Mostra tabela resumo e boxplot horizontal com cercas.
    """
    serie = df[coluna].dropna()

    # Quartis e IQR
    Q1, Q3 = serie.quantile(0.25), serie.quantile(0.75)
    IQR = Q3 - Q1
    lower, upper = Q1 - k*IQR, Q3 + k*IQR

    # Outliers
    outliers = serie[(serie < lower) | (serie > upper)]

    #  Tabela com limites
    df_limites = pd.DataFrame({
        "Q1": [Q1],
        "Q3": [Q3],
        "IQR": [IQR],
        f"Limite Inferior (k={k})": [lower],
        f"Limite Superior (k={k})": [upper],
        "Nº Outliers": [len(outliers)],
        "% Outliers": [len(outliers) / len(serie) * 100]
    }).T.rename(columns={0: "Valor"})

    display(HTML(f"<h3 style='text-align:center'>📊 Análise de Outliers (Tukey) para '{coluna}'</h3>"))
    display(df_limites.style.format({"Valor": "{:.2f}"}).set_table_styles(
        [{"selector": "table", "props": [("margin-left", "auto"), ("margin-right", "auto")]}]
    ))

    #  Boxplot horizontal
    plt.figure(figsize=(width, height))
    sns.boxplot(x=serie, orient="h", color="skyblue")
    plt.axvline(lower, color="green", linestyle="--", label="Cerca Inferior")
    plt.axvline(upper, color="red", linestyle="--", label="Cerca Superior")
    plt.title(f"Boxplot de {coluna} com Cercas de Tukey", fontsize=14, weight="bold")
    plt.legend(loc="upper right")
    plt.tight_layout()
    plt.show()

    return {"limite_inferior": lower, "limite_superior": upper, "outliers": outliers}

````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from IPython.display import display, HTML

def analisar_categorica_univariada(df, col, rename_dict=None, order=None, cores=None):
    """
    Analisa uma variável categórica:
    - Mostra tabela de frequências absoluta e relativa (%)
    - Mostra gráfico de barras com valores e percentagens
    
    Parâmetros:
    -----------
    df : DataFrame
        Base de dados
    col : str
        Nome da coluna categórica
    rename_dict : dict (opcional)
        Mapeamento de valores para rótulos legíveis (ex.: {1: "Homem", 0: "Mulher"})
    order : list (opcional)
        Ordem das categorias no gráfico
    cores : list (opcional)
        Lista de cores a aplicar às barras
    """
    
    # Criar série com rótulos mais legíveis (se houver dicionário)
    serie = df[col].map(rename_dict) if rename_dict else df[col]
    
    # Calcular frequências
    freq_table = serie.value_counts().reset_index()
    freq_table.columns = [col, 'Frequência']
    freq_table['Percentagem (%)'] = 100 * freq_table['Frequência'] / freq_table['Frequência'].sum()
    
    # Mostrar tabela formatada e centrada
    display(freq_table.style
            .format({'Percentagem (%)': '{:.2f}%'})
            .set_caption(f"📊 Distribuição de {col}")
            .set_table_attributes('style="margin:auto; width:50%; text-align:center;"'))
    display(HTML("<br><br>"))  # adiciona duas linhas em branco

    # Gráfico
    plt.figure(figsize=(8,5))
    ax = sns.countplot(x=serie, data=df, order=order, palette=cores)
    
    # Adicionar rótulos nas barras
    total = len(df)
    for p in ax.patches:
        freq = p.get_height()
        perc = 100 * freq / total
        ax.annotate(f'{freq} ({perc:.1f}%)', 
                    (p.get_x() + p.get_width() / 2., freq), 
                    ha='center', va='bottom', fontsize=11, fontweight='bold')
    
    
    
    # Ajustes de estilo
    ax.set_title(f"Distribuição de {col}", fontsize=14, weight="bold")
    ax.set_xlabel(col, fontsize=12)
    ax.set_ylabel("Número de Registos", fontsize=12)
    sns.despine()
    plt.show()
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy.stats import mannwhitneyu, pointbiserialr
from cliffs_delta import cliffs_delta
from IPython.display import display, HTML

def analisar_variavel(df, var, target="target"):
    """
    Faz análise bivariada entre uma variável numérica e uma variável target binária (0/1).
    Mostra gráficos, estatísticas e interpretação automática com hipóteses explícitas.
    """

    group0 = df.loc[df[target] == 0, var]
    group1 = df.loc[df[target] == 1, var]

    # ---------- Visualizações ----------
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    # Violin + Box
    sns.violinplot(x=df[target], y=df[var], palette="pastel", inner=None, ax=axes[0])
    sns.boxplot(x=df[target], y=df[var],
                width=0.2, boxprops={'facecolor':'none'}, ax=axes[0])
    axes[0].set_title(f"{var} vs {target} (Violin + Box)")
    axes[0].set_xlabel("Target")
    axes[0].set_ylabel(var)

    # KDE overlay
    sns.kdeplot(group0, label="Target=0", fill=True, ax=axes[1])
    sns.kdeplot(group1, label="Target=1", fill=True, ax=axes[1])
    axes[1].set_title(f"Distribuição de {var} por {target}")
    axes[1].set_xlabel(var)
    axes[1].set_ylabel("Densidade")
    axes[1].legend()

    plt.tight_layout()
    plt.show()

    # ---------- Estatísticas ----------
    desc_stats = df.groupby(target)[var].agg(['mean', 'median', 'std'])

    # Mann-Whitney U test
    u_stat, p_val = mannwhitneyu(group0, group1, alternative="two-sided")

    # Point-biserial correlation
    corr, p_corr = pointbiserialr(df[target], df[var])

    # Cliff’s Delta
    d, magnitude = cliffs_delta(group0, group1)

    # ---------- Resultados consolidados ----------
    resultados = pd.DataFrame({
        "Média Grupo 0": [group0.mean()],
        "Média Grupo 1": [group1.mean()],
        "Mediana Grupo 0": [group0.median()],
        "Mediana Grupo 1": [group1.median()],
        "Mann-Whitney U": [u_stat],
        "p-valor (U)": [p_val],
        "Correlação r": [corr],
        "p-valor (r)": [p_corr],
        "Cliff’s Delta": [d],
        "Magnitude": [magnitude]
    })

    display(resultados.style.set_caption(f"📊 Resultados Bivariados para {var}")
            .format({
                "Média Grupo 0": "{:.2f}",
                "Média Grupo 1": "{:.2f}",
                "Mediana Grupo 0": "{:.2f}",
                "Mediana Grupo 1": "{:.2f}",
                "Mann-Whitney U": "{:.2f}",
                "p-valor (U)": "{:.4f}",
                "Correlação r": "{:.3f}",
                "p-valor (r)": "{:.4f}",
                "Cliff’s Delta": "{:.3f}"
            }))

    # ---------- Interpretação automática (estilizada) ----------
    print("\n" + "="*60)
    print("📖 Interpretação Automática".center(60))
    print("="*60)

    # Mann-Whitney
    print("\n➡️ Teste **Mann-Whitney U**")
    print("   • H₀: As distribuições são iguais entre target=0 e target=1")
    print("   • H₁: As distribuições são diferentes entre target=0 e target=1")
    if p_val < 0.05:
        print(f"   ✅ Rejeitamos H₀ (p={p_val:.4f}) → **Diferença significativa entre grupos**")
    else:
        print(f"   ❌ Não rejeitamos H₀ (p={p_val:.4f}) → **Diferença não significativa**")

    # Correlação ponto-bisserial
    print("\n➡️ Teste **Correlação Ponto-Bisserial**")
    print("   • H₀: Não existe correlação linear entre variável e target (r=0)")
    print("   • H₁: Existe correlação linear entre variável e target (r≠0)")
    if p_corr < 0.05:
        print(f"   ✅ Rejeitamos H₀ (p={p_corr:.4f}) → **Correlação significativa** (r={corr:.3f})")
    else:
        print(f"   ❌ Não rejeitamos H₀ (p={p_corr:.4f}) → **Correlação não significativa** (r={corr:.3f})")

    # Cliff’s Delta
    print("\n➡️ Medida de Efeito **Cliff’s Delta (δ)**")
    print("   • Não é um teste de hipótese, mas uma medida da magnitude da diferença")
    print(f"   ⚖️ δ = {d:.3f} → **{magnitude} efeito**")
    print("="*60 + "\n")

````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency, fisher_exact
from IPython.display import display, HTML

# Função para calcular Cramér's V
def cramers_v(confusion_matrix):
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    r, k = confusion_matrix.shape
    return np.sqrt(chi2 / (n * (min(k, r)-1)))

def analisar_categorica(df, var, target="target", alpha=0.05, export=False,
                        rename_dict=None, rename_dict_target=None):
    """
    Análise bivariada entre uma variável categórica e um target binário/categórico.
    Inclui:
        - Visualização (countplot com contagens + percentuais nas barras)
        - Tabelas de contingência (absoluta e relativa)
        - Testes Qui-quadrado e Fisher (2x2)
        - Medida de efeito (Cramér’s V)
        - Resumo automático interpretado
    """

    print(f"\n📊 Resultados para {var}")
    print("-" * 70)

    df = df.copy()

    # ---------- Renomear categorias ----------
    if rename_dict:
        df[var] = df[var].map(rename_dict).fillna(df[var])

    if rename_dict_target:
        df[target] = df[target].map(rename_dict_target).fillna(df[target])

    # ---------- Visualização ----------
    plt.figure(figsize=(10, 6))
    ax = sns.countplot(x=df[var], hue=df[target], palette="pastel")

    # Calcular percentagens dentro de cada categoria (já renomeada)
    group_totals = df.groupby(df[var])[target].count()
    group_totals.index = group_totals.index.astype(str)  # garante consistência de tipos

    for container in ax.containers:
        labels = []
        for rect in container:
            height = int(rect.get_height())
            if height > 0:
                # categoria renomeada no eixo X
                cat_idx = int(rect.get_x() + rect.get_width() / 2.0 + 0.5)
                if cat_idx < len(ax.get_xticklabels()):
                    category = ax.get_xticklabels()[cat_idx].get_text()
                    # usa .get() para evitar KeyError
                    pct = 100 * height / group_totals.get(str(category), height)
                    labels.append(f"{height}\n({pct:.1f}%)")
                else:
                    labels.append("")
            else:
                labels.append("")
        ax.bar_label(container, labels=labels, label_type="edge", fontsize=9)

    # Legenda ajustada
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles, labels, title=target, loc="upper right")

    plt.title(f'{var} vs. {target}')
    plt.xlabel(var)
    plt.ylabel("Contagem")
    plt.tight_layout()
    plt.show()

    # ---------- Tabelas ----------
    contingency = pd.crosstab(df[var], df[target])

    display(HTML("<div style='text-align:center'>"))
    display(contingency.style
            .set_caption("📌 Tabela de Contingência (absoluta)")
            .set_table_attributes('style="margin:auto"')
            .background_gradient(axis=None, cmap="Blues"))

    contingency_pct = contingency.apply(lambda x: x / x.sum(), axis=1)
    display(contingency_pct.style
            .set_caption("📌 Tabela de Contingência (% por linha)")
            .set_table_attributes('style="margin:auto"')
            .format("{:.2%}")
            .background_gradient(axis=None, cmap="Greens"))
    display(HTML("</div>"))

    # ---------- Testes de hipóteses ----------
    chi2, p, dof, expected = chi2_contingency(contingency)

    fisher_res = None
    if contingency.shape == (2, 2):
        oddsratio, p_fisher = fisher_exact(contingency)
        fisher_res = (oddsratio, p_fisher)

    # ---------- Medida de efeito ----------
    v = cramers_v(contingency)
    if v < 0.1:
        interpret = "negligenciável"
    elif v < 0.3:
        interpret = "pequena"
    elif v < 0.5:
        interpret = "moderada"
    else:
        interpret = "forte"

    # ---------- Resumo automático ----------
    print("\n" + "="*60)
    print(f"📖 Interpretação Automática de {var}".center(60))
    print("="*60)

    print("\n➡️ Teste Qui-quadrado de independência")
    print("   • H₀: Não existe associação entre as variáveis (independência)")
    print("   • H₁: Existe associação entre as variáveis (dependência)")
    if p < alpha:
        print(f"   ✅ Rejeitamos H₀ (p={p:.4f}) → **Associação significativa**")
    else:
        print(f"   ❌ Não rejeitamos H₀ (p={p:.4f}) → **Sem evidência de associação**")

    if fisher_res:
        oddsratio, p_fisher = fisher_res
        print("\n➡️ Teste Exato de Fisher (2x2)")
        print("   • H₀: As probabilidades de ocorrência do target são iguais entre categorias")
        print("   • H₁: As probabilidades diferem entre categorias")
        if p_fisher < alpha:
            print(f"   ✅ Rejeitamos H₀ (p={p_fisher:.4f}) → **Associação significativa** (OR={oddsratio:.2f})")
        else:
            print(f"   ❌ Não rejeitamos H₀ (p={p_fisher:.4f}) → **Sem evidência de associação** (OR={oddsratio:.2f})")

    print("\n➡️ Medida de efeito: Cramér’s V")
    print(f"   • V = {v:.3f} → **{interpret} associação**")
    print("="*60 + "\n")

    # ---------- Exportação ----------
    if export:
        resultados = {
            "Variável": var,
            "Chi2": chi2,
            "p-valor Qui2": p,
            "Cramér's V": v,
            "Força": interpret
        }
        if fisher_res:
            resultados.update({"OddsRatio": oddsratio, "p-valor Fisher": p_fisher})
        return resultados
````

### 🔎 Justificação Metodológica — Correlação (Spearman) e VIF

#### 1. Porque utilizámos a correlação de Spearman (e não a de Pearson)
A análise de correlação foi realizada com o coeficiente **de Spearman** em vez do coeficiente **de Pearson**, pelas seguintes razões:

- **Suposições menos restritivas:**  
  O coeficiente de Pearson assume linearidade e distribuição normal das variáveis.  
  Já Spearman, por ser uma correlação baseada em **ranks**, não depende da normalidade nem da linearidade estrita.  

- **Robustez a outliers e assimetrias:**  
  Muitas das variáveis numéricas do dataset (ex.: colesterol, pressão arterial, oldpeak) apresentam **assimetria** e **outliers identificados nas análises univariadas**.  
  O coeficiente de Spearman é mais robusto a esses desvios, tornando-se mais adequado.  

- **Natureza dos dados clínicos:**  
  Variáveis fisiológicas nem sempre têm relação linear com a presença de doença, mas podem apresentar relações **monotónicas** (à medida que a variável aumenta, o risco pode aumentar ou diminuir, mas não necessariamente de forma linear).  
  O Spearman capta essas tendências monotónicas, ao contrário do Pearson.

**Conclusão:** A escolha do Spearman garante uma medida de associação mais fiável e robusta para este tipo de dados clínicos.

---

#### 2. Porque complementámos com a análise de VIF
Após a análise da matriz de correlação de Spearman, foi efetuada a análise do **Variance Inflation Factor (VIF)** com o objetivo de avaliar a **multicolinearidade** entre as variáveis explicativas.

- **Limitação da matriz de correlação:**  
  A matriz de correlação apenas avalia relações **bivariadas** (entre dois pares de variáveis).  
  Uma variável pode não ter correlação alta com nenhuma outra em particular, mas ainda assim ser explicada por uma **combinação linear de várias variáveis**.  

- **VIF como análise multivariada:**  
  O VIF mede quanto a variância de cada variável é inflacionada pela presença das outras no modelo.  
  Valores elevados de VIF (≥ 5 ou ≥ 10, consoante a literatura) indicam **colinearidade preocupante**, o que pode:  
  - Inflacionar erros-padrão,  
  - Instabilizar coeficientes de regressão,  
  - Dificultar a interpretação da contribuição individual de cada variável.

- **Importância para modelagem:**  
  Em modelos preditivos (como regressões logísticas ou lineares), a multicolinearidade compromete a **estabilidade e interpretabilidade**.  
  A análise do VIF assegura que as variáveis retidas são estatisticamente consistentes.

**Conclusão:**  
A matriz de correlação de Spearman permitiu identificar **associações monotónicas e robustas** entre variáveis.  
O VIF complementou esta análise ao detetar redundâncias **multivariadas** no conjunto completo, garantindo que as variáveis selecionadas para os modelos não apresentam colinearidade problemática.

````python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd
import statsmodels.api as sm
from IPython.display import display

def calcular_vif_final(df, numeric_var, categoric_var):
    # 1. Selecionar numéricas
    num_data = df[numeric_var].dropna()
    
    # 2. Criar dummies para categóricas
    cat_data = pd.get_dummies(df[categoric_var], drop_first=True)
    
    # 3. Juntar numéricas + categóricas
    X_all = pd.concat([num_data, cat_data], axis=1)
    X_all = sm.add_constant(X_all)
    
    # 4. Calcular VIF
    vif_all = pd.DataFrame()
    vif_all["Variável"] = X_all.columns
    vif_all["VIF"] = [variance_inflation_factor(X_all.values, i) for i in range(X_all.shape[1])]
    
    # 5. Mostrar tabela estilizada
    display(
        vif_all.sort_values("VIF", ascending=False).style
        .set_caption(" Fatores de Inflação da Variância (VIF) — Todas as Variáveis")
        .format({"VIF": "{:.2f}"})
        .background_gradient(subset="VIF", cmap="coolwarm")
        .set_table_styles(
            [
                {"selector": "caption", "props": [("font-size", "16px"), ("font-weight", "bold"), ("color", "#9EA3B1")]},
                {"selector": "th", "props": [("background-color", "#1F1F1F"), ("font-weight", "bold")]}
            ]
        )
    )
    
    return vif_all.sort_values("VIF", ascending=False)

````

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_validate
from sklearn.preprocessing import RobustScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    roc_auc_score,
    roc_curve
)

# ======================
# 1. Definir X e y
# ======================
X = df.drop(columns=["target"])
y = df["target"]

# ======================
# 2. Identificar colunas
# ======================
numeric_features = ["age", "trtbps", "chol", "thalach", "oldpeak"]
categorical_features = ["sex", "cp", "fbs", "rest_ecg", "exang", "slope", "ca", "thal"]

# ======================
# 3. Pré-processamento
# ======================
preprocessor = ColumnTransformer(
    transformers=[
        ("num", RobustScaler(), numeric_features),
        ("cat", OneHotEncoder(drop="first"), categorical_features)
    ]
)

# ======================
# 4. Criar pipeline
# ======================
pipe = Pipeline([
    ("preprocessor", preprocessor),
    ("log_reg", LogisticRegression(max_iter=1000, random_state=19))
])

# ======================
# 5. Divisão treino/teste
# ======================
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=3, stratify=y
)

import optuna
import numpy as np
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

# =============================
# Função objetivo do Optuna
# =============================
def objective(trial):
    model_type = trial.suggest_categorical("model", [
        "logreg", "rf", "xgb", "lgbm", "catboost", "knn", "svc"
    ])

    if model_type == "logreg":
        C = trial.suggest_loguniform("C", 1e-3, 1e3)
        penalty = trial.suggest_categorical("penalty", ["l1", "l2"])
        solver = "liblinear" if penalty == "l1" else "lbfgs"
        model = LogisticRegression(
            C=C, penalty=penalty, solver=solver, max_iter=500
        )

    elif model_type == "rf":
        n_estimators = trial.suggest_int("n_estimators", 100, 500)
        max_depth = trial.suggest_int("max_depth", 3, 15)
        min_samples_split = trial.suggest_int("min_samples_split", 2, 10)
        min_samples_leaf = trial.suggest_int("min_samples_leaf", 1, 5)
        model = RandomForestClassifier(
            n_estimators=n_estimators,
            max_depth=max_depth,
            min_samples_split=min_samples_split,
            min_samples_leaf=min_samples_leaf,
            random_state=3
        )

    elif model_type == "xgb":
        n_estimators = trial.suggest_int("n_estimators", 100, 500)
        max_depth = trial.suggest_int("max_depth", 3, 10)
        learning_rate = trial.suggest_loguniform("learning_rate", 0.01, 0.3)
        subsample = trial.suggest_uniform("subsample", 0.6, 1.0)
        colsample_bytree = trial.suggest_uniform("colsample_bytree", 0.6, 1.0)
        model = XGBClassifier(
            n_estimators=n_estimators,
            max_depth=max_depth,
            learning_rate=learning_rate,
            subsample=subsample,
            colsample_bytree=colsample_bytree,
            random_state=3,
            use_label_encoder=False,
            eval_metric="logloss"
        )

    elif model_type == "lgbm":
        n_estimators = trial.suggest_int("n_estimators", 100, 500)
        max_depth = trial.suggest_int("max_depth", -1, 15)  # -1 = sem limite
        learning_rate = trial.suggest_loguniform("learning_rate", 0.01, 0.3)
        num_leaves = trial.suggest_int("num_leaves", 15, 50)
        model = LGBMClassifier(
            n_estimators=n_estimators,
            max_depth=max_depth,
            learning_rate=learning_rate,
            num_leaves=num_leaves,
            random_state=3
        )

    elif model_type == "catboost":
        depth = trial.suggest_int("depth", 3, 10)
        learning_rate = trial.suggest_loguniform("learning_rate", 0.01, 0.3)
        iterations = trial.suggest_int("iterations", 100, 500)
        model = CatBoostClassifier(
            depth=depth,
            learning_rate=learning_rate,
            iterations=iterations,
            random_state=3,
            verbose=0
        )

    elif model_type == "knn":
        n_neighbors = trial.suggest_int("n_neighbors", 3, 15)
        metric = trial.suggest_categorical("metric", ["euclidean", "manhattan", "minkowski"])
        model = KNeighborsClassifier(
            n_neighbors=n_neighbors, metric=metric
        )

    elif model_type == "svc":
        C = trial.suggest_loguniform("C", 1e-3, 1e3)
        kernel = trial.suggest_categorical("kernel", ["linear", "rbf", "poly"])
        gamma = trial.suggest_loguniform("gamma", 1e-4, 1e-1) if kernel != "linear" else "scale"
        model = SVC(
            C=C, kernel=kernel, gamma=gamma, probability=True
        )

    # =============================
    # Validação cruzada estratificada
    # =============================
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=3)
    scores = cross_val_score(model, X_train, y_train, cv=cv, scoring="roc_auc")
    
    return np.mean(scores)

# =============================
# Executar estudo
# =============================
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=100, n_jobs=-1)  # 100 iterações

print("Melhor modelo:", study.best_trial.params["model"])
print("Melhores parâmetros:", study.best_trial.params)
print("Melhor ROC AUC:", study.best_value)
````

````python
import numpy as np

class ThresholdClassifier:
    def __init__(self, pipeline, threshold=0.5):
        self.pipeline = pipeline
        self.threshold = threshold
    
    def fit(self, X, y):
        self.pipeline.fit(X, y)
        return self
    
    def predict_proba(self, X):
        return self.pipeline.predict_proba(X)
    
    def predict(self, X):
        probs = self.predict_proba(X)[:, 1]  # probabilidade da classe positiva
        return (probs >= self.threshold).astype(int)
````
